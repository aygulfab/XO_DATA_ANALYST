```python
df_leads.created_at = pd.to_datetime(df_leads.created_at, errors = 'coerce')
df_transactions.created_at = pd.to_datetime(df_transactions.created_at, errors = 'coerce') 
df_clients.created_at = pd.to_datetime(df_clients.created_at, errors = 'coerce') 
```

```python
#обрезаем датасеты по дням отчета
DATE_START = pd.to_datetime(DATE_START)
DATE_END = pd.to_datetime(DATE_END)
```

```python
df_leads_select = df_leads[(DATE_START <= df_leads['created_at']) & (df_leads['created_at'] <= DATE_END)]
```

```python
result = [] #итоговая таблица
for manager in df_managers['manager_id']:
    for source in df_leads_select['d_utm_source'].unique():
        leads = df_leads_select[(df_leads_select.d_utm_source == source) & (df_leads_select.l_manager_id == manager)] #количество заявок
        leads_noclient = 0 #количество мусорных заявок
        leads_new = 0  #количество новых заявок
        leads_tocustomer = 0  #количество покупателей
        leads_tocustomer_new = 0  #количество новых покупателей
        revenue = 0  #доход от покупок новых покупателей
        for client, day in zip(leads.l_client_id, leads.created_at):        
            if len(df_clients[(df_clients.client_id == client) & (df_clients.l_manager_id == manager)]) == 0:
                leads_noclient += 1  
            if (len(df_leads[(df_leads.l_client_id == client) & (df_leads.created_at < day)]) == 0) & (len(df_transactions[(df_transactions.l_client_id == client) & (df_transactions.created_at < day)]) == 0):
                leads_new += 1    
            day2=day + datetime.timedelta(days=7) 
            condition1 = ((day <= df_transactions.created_at) & (df_transactions.created_at <= day2)) #купил в течение недели
            condition2 = (df_transactions.l_client_id == client) #текущий клиент
            condition3 = ((df_transactions.created_at < day)) #не покупал раньше
            if len(df_transactions[condition1 & condition2] != 0):
                leads_tocustomer += 1
                if len(df_transactions[condition2 & condition3]) == 0:
                    leads_tocustomer_new += 1
                    transactions = df_transactions[condition1 & condition2] #вытаскиваем транзакции за неделю
                    revenue += transactions.m_real_amount.sum()
        result.append([manager, source, len(leads), leads_noclient, leads_new, leads_tocustomer, leads_tocustomer_new, revenue])
```

```python
print(*result, sep = "\n")
```

```python
df = pd.DataFrame(result, columns = ['manager','канал привлечения заявки','количество заявок','количество мусорных заявок','количество новых заявок','количество покупателей', 'количество новых покупателей', 'доход от покупок новых покупателей'])
df
```
